```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
july_2014 = "Data/2014-07 - Citi Bike trip data.csv"
```

```python
# Read our Data file with the pandas library
july_2014_df = pd.read_csv(july_2014, encoding="ISO-8859-1")
```

```python
# Show just the header
july_2014_df.head()
```

```python
# Delete columns we don't care about
del july_2014_df['gender']
july_2014_df.head()
```

```python
# Delete columns we don't care about
del july_2014_df['birth year']
july_2014_df.head()
```

```python
# Delete columns we don't care about
del july_2014_df['bikeid']
july_2014_df.head()
```

```python
# Using GroupBy in order to separate the data into fields according to "usertype" values
grouped_july_2014_df = july_2014_df.groupby(['start station name'])

# Visualize
grouped_july_2014_df.count().head(10)
```

```python
# Converting a GroupBy object into a DataFrame
new_july_2014_df = pd.DataFrame(
    grouped_july_2014_df["usertype"].count())
new_july_2014_df.head(10)
```

```python
# Export file as a CSV, without the Pandas index, but with the header
new_july_2014_df.to_csv("Data/july_2014.csv", header=True)
```
